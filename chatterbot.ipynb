```python
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
import os
import chatterbot_corpus

bot=ChatBot('solete')
training=ChatterBotCorpusTrainer(bot)

# Construct the absolute path to the English corpus data
corpus_data_path = os.path.join(os.path.dirname(chatterbot_corpus.__file__), 'data', 'english')

training.train(corpus_data_path)
print(bot.get_response('hello'))
print('\n Aske me something::')
flag=True
while flag:
  userinput=input('\nYou:')
  if userinput=='done':
    flag=False
    print('Bot: Bye Bye')
  else:
    response=bot.get_response(userinput)
    print('\n\nBot:',response)
```
