```python
# Step 1: Movie data
movies = {
    "Action": ["Mad Max", "John Wick"],
    "Comedy": ["The Mask", "Superbad"],
    "Drama": ["The Godfather", "Shawshank Redemption"]
}

# Step 2: Create a reverse mapping: movie -> genre
movie_to_genre = {}
for genre, titles in movies.items():
    for title in titles:
        movie_to_genre[title.lower()] = genre

# Step 3: User inputs a movie name
movie_name = input("Enter a movie name: ").strip().lower()

# Step 4: Recommend similar movies
if movie_name in movie_to_genre:
    genre = movie_to_genre[movie_name]
    similar_movies = [m for m in movies[genre] if m.lower() != movie_name]
    print(f"Genre: {genre}")
    print("You might also like:", similar_movies)
else:
    print("Movie not found in database.")
```
